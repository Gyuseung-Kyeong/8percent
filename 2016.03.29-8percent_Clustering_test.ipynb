```python
import pandas as pd
import numpy as np
import math
```

```python
df_after_cnt_vec=pd.read_excel('df_after_cnt_vec.xlsx')
```

```python
df_after_cnt_vec.head()
```

```python
df_after_cnt_vec_with_index=df_after_cnt_vec.set_index('index')
df_after_cnt_vec_with_index.head()
```

```python
import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
from sklearn.cluster import AgglomerativeClustering
K = 20
X = df_after_cnt_vec_with_index.values
hi_cluster = AgglomerativeClustering(n_clusters=K, affinity='cosine', linkage='average')
hi_cluster.fit(X)
```

```python
hi_cluster1 = pd.DataFrame( [[a,b] for a,b in zip(df_after_cnt_vec_with_index.index, hi_cluster.labels_)], \
                         columns=['index','cluster-id'], \
                         index=df_after_cnt_vec_with_index.index )
hi_cluster1.groupby(['cluster-id']).groups
```

```python
hi_cluster1
```

```python
hi_cluster1.groupby('cluster-id').count()
```

```python
df_with_index_cluster=pd.merge(df_after_cnt_vec,hi_cluster1,how='left',on='index')
df_with_index_cluster.head()
```

```python
df_mean=pd.DataFrame(columns=df_with_index_cluster.columns)
df_std=pd.DataFrame(columns=df_with_index_cluster.columns)
df_count=pd.DataFrame(columns=df_with_index_cluster.columns)
for i in range(0,20):
    mean_temp=pd.DataFrame(df_with_index_cluster[df_with_index_cluster['cluster-id']==i].mean(),columns=[i]).transpose()
    std_temp=pd.DataFrame(df_with_index_cluster[df_with_index_cluster['cluster-id']==i].std(),columns=[i]).transpose()
    count_temp=pd.DataFrame(df_with_index_cluster[df_with_index_cluster['cluster-id']==i].count(),columns=[i]).transpose()
    df_mean=pd.concat([df_mean,mean_temp])
    df_std=pd.concat([df_std,std_temp])
    df_count=pd.concat([df_count,count_temp])
```

```python
del df_mean['index']
del df_std['index']
```

```python
df_mean_count=pd.concat([df_mean,df_count['index']],axis=1)
df_std_count=pd.concat([df_std,df_count['index']],axis=1)
```

```python
df_mean_count.rename(columns={'index':'count'},inplace=True)
df_std_count.rename(columns={'index':'count'},inplace=True)
```

```python
df_std_count.head(3)
```

```python
df_mean_count['mean']=1
df_mean_count['std']=0
df_std_count['mean']=0
df_std_count['std']=1
```

```python
df_summary=pd.concat([df_mean_count,df_std_count],axis=0)
```

```python
df_summary_temp=df_summary.reset_index()
df_summary_temp.sort(['index','std']).to_excel('summary.xlsx','summary')
```
