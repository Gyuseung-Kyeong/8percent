```python
import pandas as pd
import xlrd
import re
```

```python
def raw_data_preprocessing(file):
    raw_data0=pd.read_excel(file)
    raw_data0_0 = raw_data0[raw_data0['심사총론'].notnull()]
    raw_data1=raw_data0_0[raw_data0_0['RTI']!='None']
    raw_data2=pd.DataFrame()
    raw_data2['url_number']=raw_data1['url_number']
    raw_data2['채권명']=raw_data1['채권명'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['투자자모집시작일']=raw_data1['투자자모집시작일']
    raw_data2['모집금액']=raw_data1['모집금액'].apply(
        lambda x: int(x.replace(',',''))/100
        )
    raw_data2['안심펀드']=raw_data1['안심펀드'].apply(
        lambda x: 0 if x=='None' else float(x[0:-1])
        )
    raw_data2['연금리']=raw_data1['연금리'].apply(
        lambda x: float(x[0:-1])
        )
    raw_data2['투자현황']=raw_data1['투자현황']
    raw_data2['대출목적']=raw_data1['대출목적']
    raw_data2['상품']=raw_data1['상품']
    raw_data2['만기']=raw_data1['만기'].apply(
        lambda x: int(x[0:-2])
        )
    raw_data2['대출금액']=raw_data1['대출금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )
    raw_data2['상환방식']=raw_data1['상환방식']
    raw_data2['월상환액']=raw_data1['월상환액'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['8percent신용등급']=raw_data1['8percent신용등급']
    raw_data2['8percent예측불량률']=raw_data1['8percent예측불량률'].apply(
        lambda x: float(x[0:-1])
        )
    raw_data2['KCB신용등급']=raw_data1['KCB신용등급'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['KCB신용점수']=raw_data1['KCB신용등급'].apply(
        lambda x: re.findall("\d+", x)[1]
        )
    raw_data2['KCB예측불량률']=raw_data1['KCB예측불량률'].apply(
        lambda x: float(x[0:-1])
        )
    raw_data2['주요사항']=raw_data1['주요사항']
    raw_data2['담보건수']=raw_data1['담보건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['담보금액']=raw_data1['담보금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )

    raw_data2['은행보험학자금건수']=raw_data1['은행보험학자금건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['은행보험학자금금액']=raw_data1['은행보험학자금금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )

    raw_data2['저축은행캐피탈카드론건수']=raw_data1['저축은행캐피탈카드론건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['저축은행캐피탈카드론금액']=raw_data1['저축은행캐피탈카드론금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )

    raw_data2['현금서비스건수']=raw_data1['현금서비스건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['현금서비스금액']=raw_data1['현금서비스금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )

    raw_data2['현금서비스건수']=raw_data1['현금서비스건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['현금서비스금액']=raw_data1['현금서비스금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )
    raw_data2['대부건수']=raw_data1['대부건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['대부금액']=raw_data1['대부금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )

    raw_data2['P2P건수']=raw_data1['P2P건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['P2P금액']=raw_data1['P2P금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )

    raw_data2['총합건수']=raw_data1['총합건수'].apply(
        lambda x: re.findall("\d+", x)[0]
        )
    raw_data2['총합금액']=raw_data1['총합금액'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )
    raw_data2['월소득']=raw_data1['월소득'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )
    raw_data2['RTI']=raw_data1['RTI'].apply(
        lambda x: float(x[0:-1])
        )
    raw_data2['월가처분소득']=raw_data1['월가처분소득'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )
    raw_data2['LTI']=raw_data1['LTI'].apply(
        lambda x: float(x[0:-1])
        )
    raw_data2['직장']=raw_data1['직장']
    raw_data2['업종']=raw_data1['업종']
    raw_data2['근무기간']=raw_data1['근무기간'].apply(
        lambda x: float(x[0:-1])
        )
    raw_data2['고용형태']=raw_data1['고용형태']
    raw_data2['연소득']=raw_data1['연소득'].apply(
        lambda x: int(x[0:-2].replace(',',''))/100
        )
    raw_data2['심사총론']=raw_data1['심사총론']
    return raw_data2.to_excel('preprocessed_data.xlsx','preprocessed_data')
```
